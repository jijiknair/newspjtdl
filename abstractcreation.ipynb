```python
#abstract creation from an article
!pip install sumy
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lex_rank import LexRankSummarizer

# Sample article text
article_text = """
    
In recent years, technology has revolutionized the field of education, transforming the way students learn and teachers teach. From interactive whiteboards to online learning platforms, technology has made education more accessible, engaging, and personalized than ever before.

One of the key impacts of technology on education is the shift towards digital classrooms. With the widespread availability of computers, tablets, and smartphones, students can now access educational resources and learning materials from anywhere with an internet connection. This has led to the rise of online learning platforms, which offer courses on a wide range of subjects, from mathematics to literature.

Furthermore, technology has enabled teachers to adopt innovative teaching methods that cater to different learning styles. For example, interactive whiteboards allow teachers to create dynamic presentations, videos, and interactive quizzes to engage students in the learning process. Additionally, educational apps and games make learning fun and accessible, encouraging students to explore new concepts and ideas in a hands-on way.
"""

# Initialize the parser and tokenizer
parser = PlaintextParser.from_string(article_text, Tokenizer("english"))

# Initialize LexRank summarizer
summarizer = LexRankSummarizer()

# Generate summary
summary = summarizer(parser.document, sentences_count=2)  # You can adjust sentences_count as needed

# Print the summary
print("Summary:")
for sentence in summary:
    print(sentence)
```
